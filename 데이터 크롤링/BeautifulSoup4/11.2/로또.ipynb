```python
### 회차 상호명 구분 주소
### 1000~1038회 pandas
```

```python
#포스트 인자를 프로토콜에 넣어서 전달
#get url 안에 들어가 있음
```

```python
data3=[]
for i in range(1000,1039):
    data5 = {'method':'topStore',
             'nowPage':'1',
             'rankNo':'',
             'gameNo':'5133',
             'drwNo':i,
             'schKey':'all',
             'schVal':''}
    
    req2=requests.post(url2,data=data5)
    html2=req2.text
    soup2=BeautifulSoup(html2,'html.parser')
    
    
    name2=soup2.select('#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr> td:nth-child(2)')
    division2=soup2.select('#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr > td:nth-child(3)')
    address2=soup2.select('#article > div:nth-child(2) > div > div:nth-child(4) > table > tbody > tr> td:nth-child(4)')
    

    for j in range(len(name2)):
        temp=[]
        temp.append(i)
        temp.append(name2[j].text)
        temp.append(division2[j].text.strip())
        temp.append(address2[j].text)
        data3.append(temp)
```

```python
df=pd.DataFrame(data3,columns=['회차','상호명','구분','소재지'])
```
